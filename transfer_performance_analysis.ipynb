# Import Required Libraries

```python
import pandas as pd
```

# Load Transfer Data

```python
transfers = pd.read_csv('datasets/Football Data from Transfermarkt/transfers.csv')
transfers = transfers[['player_name', 'transfer_season', 'market_value_in_eur']]
```

# Load Player Performance Data

```python
players = pd.read_csv('datasets/Football Players Datasets 2015 - 2024/Latest Football  Players 2024 Data.csv')
players = players[['Players', 'Seasons', 'Goals', 'Seasons Ratings']]
```

# Merge Datasets on Player Name

```python
merged = pd.merge(transfers, players, left_on='player_name', right_on='Players', how='left')
```

```python
def get_start_year(season):
    if pd.isna(season):
        return None
    if isinstance(season, str) and '/' in season:
        return int(season.split('/')[0])
    else:
        return int(season)

merged['transfer_start_year'] = merged['transfer_season'].apply(lambda x: int('20' + x.split('/')[0]) if isinstance(x, str) and '/' in x else (int(x) if not pd.isna(x) else None))
merged['season_start_year'] = merged['Seasons'].apply(get_start_year)
```

# Calculate Previous Season Goals

```python
filtered = merged[(merged['season_start_year'].notna()) & (merged['transfer_start_year'].notna()) & (merged['season_start_year'] == merged['transfer_start_year'] - 1)]
result = filtered[['player_name', 'transfer_season', 'market_value_in_eur', 'Goals', 'Seasons Ratings']].copy()
result.columns = ['playername', 'transfer season', 'market valuation', 'number of goals scored in the previous season', 'season rating for the previous season']
```

# Create and Save Output CSV

```python
result.to_csv('transfer_performance.csv', index=False)
```
