# Player Name Matching Between Datasets

This notebook reads two football player CSV datasets and matches player names between them.

## 1. Import Required Libraries
Import pandas and any other necessary libraries for data manipulation and matching.

```python
import pandas as pd
import numpy as np
import re
import unicodedata
```

## 2. Load CSV Files
Read the two CSV files into pandas DataFrames.

```python
# Load the CSV files into DataFrames
players_2024_df = pd.read_csv('../datasets/Football Players Datasets 2015 - 2024/Latest Football  Players 2024 Data.csv')
transfermarkt_players_df = pd.read_csv('../datasets/Football Data from Transfermarkt/players.csv')
```

## 3. Preview DataFrames
Display the first few rows of each DataFrame to understand their structure, focusing on the 'Players' and 'name' columns.

```python
# Preview the first few rows of each DataFrame
print("Players 2024 DataFrame:")
print("Number of players in 2024 dataset:", len(players_2024_df))
display(players_2024_df[['Players']].head())

print("Transfermarkt Players DataFrame:")
print("Number of players in Transfermarkt dataset:", len(transfermarkt_players_df))
display(transfermarkt_players_df[['name']].head())
```

## 4. Match Player Names Across DataFrames
Perform a merge operation to match player names from the 'Players' column in the first DataFrame with the 'name' column in the second DataFrame.

```python
# Merge DataFrames on normalized player names
matched_df_plain = pd.merge(
    players_2024_df,
    transfermarkt_players_df,
    left_on='Players',
    right_on='name',
    how='inner',
    suffixes=('_2024', '_tm')
)
print(f"Number of matched players: {len(matched_df_plain)}")
```

## 5. Display Matched Results
Show the resulting DataFrame with matched player names and relevant columns from both datasets.

```python
# Display the matched results
# Show selected columns for clarity
columns_to_show = [
    'Players', 'Teams', 'Seasons', 'Matches', 'Goals', 'Assists', 'Seasons Ratings',
    'name', 'player_id'
]
display(matched_df_plain[columns_to_show].head(20))
```

## 6. Display Non-Matched Player Names
Show the player names from the first CSV that were not matched to any name in the second CSV.

```python
# Find and display non-matched player names from the first CSV
matched_names_plain = set(matched_df_plain['Players'])
all_names = set(players_2024_df['Players'])
non_matched_names_plain = sorted(all_names - matched_names_plain)

print(f"Number of non-matched player names: {len(non_matched_names_plain)}")
for name in non_matched_names_plain:
    print(name)
```

## 7. Add a function to Normalize Player Names
Clean and normalize the player names in both DataFrames to improve matching accuracy.

```python
# Function to normalize player names
def normalize_name(name):
    if pd.isnull(name):
        return ""
    # Lowercase, strip, remove special characters
    name = name.lower().strip()
    name = re.sub(r'[^a-z0-9 ]', '', name)
    name = re.sub(r'\s+', ' ', name)
    return name

# Apply normalization to both DataFrames
players_2024_df['Players_normalized'] = players_2024_df['Players'].apply(normalize_name)
transfermarkt_players_df['name_normalized'] = transfermarkt_players_df['name'].apply(normalize_name)
```

## 8. Match Player Names Across DataFrames After Normalization
Perform a merge operation to match player names from the 'Players' column in the first DataFrame with the 'name' column in the second DataFrame.

```python
# Merge DataFrames on normalized player names
matched_df_normalized = pd.merge(
    players_2024_df,
    transfermarkt_players_df,
    left_on='Players_normalized',
    right_on='name_normalized',
    how='inner',
    suffixes=('_2024', '_tm')
)
print(f"Number of matched players: {len(matched_df_normalized)}")
```

## 9. Display Matched Results After Normalization
Show the resulting DataFrame with matched player names and relevant columns from both datasets.

```python
# Display the matched results
# Show selected columns for clarity
columns_to_show = [
    'Players', 'Teams', 'Seasons', 'Matches', 'Goals', 'Assists', 'Seasons Ratings',
    'name', 'player_id'
]
display(matched_df_normalized[columns_to_show].head(20))
```

## 10. Display Non-Matched Player Names after Normalization
Show the player names from the first CSV that were not matched to any name in the second CSV.

```python
# Find and display non-matched player names from the first CSV
matched_names_normalized = set(matched_df_normalized['Players'])
all_names = set(players_2024_df['Players'])
non_matched_names_normalized = sorted(all_names - matched_names_normalized)

print(f"Number of non-matched player names: {len(non_matched_names_normalized)}")
for name in non_matched_names_normalized:
    print(name)
```

## 11. Improve the Normalization Function Treating Special Characters

Convert special characters (such as vowels with accents) to teh corresponding character as part of the 
normalization, and run matching again

```python
# Function to normalize player names
def normalize_name(name):
    if pd.isnull(name):
        return ""
    # Lowercase, strip, remove special characters
    name = name.lower().strip()
    # Remove accents: decompose unicode and keep only ASCII characters
    name = unicodedata.normalize('NFD', name)
    name = ''.join(c for c in name if unicodedata.category(c) != 'Mn')
    # Remove special characters
    name = re.sub(r'[^a-z0-9 ]', '', name)
    name = re.sub(r'\s+', ' ', name)
    return name

# Apply normalization to both DataFrames
players_2024_df['Players_normalized'] = players_2024_df['Players'].apply(normalize_name)
transfermarkt_players_df['name_normalized'] = transfermarkt_players_df['name'].apply(normalize_name)

# Merge DataFrames on normalized player names
matched_df_normalized = pd.merge(
    players_2024_df,
    transfermarkt_players_df,
    left_on='Players_normalized',
    right_on='name_normalized',
    how='inner',
    suffixes=('_2024', '_tm')
)
print(f"Number of matched players: {len(matched_df_normalized)}")

# Find and display non-matched player names from the first CSV
matched_names_normalized = set(matched_df_normalized['Players'])
all_names = set(players_2024_df['Players'])
non_matched_names_normalized = sorted(all_names - matched_names_normalized)

print(f"Number of non-matched player names: {len(non_matched_names_normalized)}")
for name in non_matched_names_normalized:
    print(name)
```

## 12. Display Names Matched Only After Normalization
Show the player names that were matched in the normalization step but not in the plain step.

```python
# Find names matched only after normalization
matched_names_plain = set(matched_df_plain['Players'])
matched_names_normalized = set(matched_df_normalized['Players'])

matched_only_after_normalization = sorted(matched_names_normalized - matched_names_plain)

print(f"Number of player names matched only after normalization: {len(matched_only_after_normalization)}")
for name in matched_only_after_normalization:
    print(name)
```
